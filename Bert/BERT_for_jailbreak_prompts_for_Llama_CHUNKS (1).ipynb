```python
# Добавляем метки (проверка наличия столбца 'label' перед добавлением)
def add_label(df, label_value):
    if 'label' not in df.columns:
        df['label'] = label_value
    return df

# Добавляем метки
jailbreak_prompts_2023_05_07 = add_label(jailbreak_prompts_2023_05_07, 1)
jailbreak_prompts_2023_12_25 = add_label(jailbreak_prompts_2023_12_25, 1)
regular_prompts_2023_05_07 = add_label(regular_prompts_2023_05_07, 0)
regular_prompts_2023_12_25 = add_label(regular_prompts_2023_12_25, 0)

# Объединяем все датасеты в один
all_data = pd.concat([jailbreak_prompts_2023_05_07, jailbreak_prompts_2023_12_25, regular_prompts_2023_05_07, regular_prompts_2023_12_25], ignore_index=True)

# Сохраняем объединённый датасет для дальнейшего использования
all_data.to_csv('/content/combined_prompts.csv', index=False)

print("Combined data sample:")
display(all_data.head())
```

```python
from transformers import BertTokenizer
import torch

# Загружаем токенизатор
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Функция для разбиения текста на чанки
def chunk_text(text, max_length=512):
    # Токенизируем текст
    tokens = tokenizer.encode(text, truncation=False)

    # Разбиваем токены на чанки
    chunks = [tokens[i:i + max_length] for i in range(0, len(tokens), max_length)]

    return chunks

# Применяем к каждому запросу из обучающего и тестового набора
X_train_chunks = [chunk_text(prompt) for prompt in X_train.tolist()]
X_test_chunks = [chunk_text(prompt) for prompt in X_test.tolist()]

# Теперь у нас есть чанки для каждого запроса. Мы можем токенизировать их по частям.
train_input_ids = []
train_attention_mask = []
for chunks in X_train_chunks:
    for chunk in chunks:
        tokenized_chunk = tokenizer.encode_plus(chunk, padding='max_length', truncation=True, max_length=512, return_tensors='pt')
        train_input_ids.append(tokenized_chunk['input_ids'])
        train_attention_mask.append(tokenized_chunk['attention_mask'])

test_input_ids = []
test_attention_mask = []
for chunks in X_test_chunks:
    for chunk in chunks:
        tokenized_chunk = tokenizer.encode_plus(chunk, padding='max_length', truncation=True, max_length=512, return_tensors='pt')
        test_input_ids.append(tokenized_chunk['input_ids'])
        test_attention_mask.append(tokenized_chunk['attention_mask'])

# Преобразуем в DataFrame
train_df = pd.DataFrame({
    'input_ids': [i.squeeze(0).tolist() for i in train_input_ids],
    'attention_mask': [i.squeeze(0).tolist() for i in train_attention_mask]
})

test_df = pd.DataFrame({
    'input_ids': [i.squeeze(0).tolist() for i in test_input_ids],
    'attention_mask': [i.squeeze(0).tolist() for i in test_attention_mask]
})

print(train_df.head())  # Проверка результата
```

```python
from transformers import BertTokenizer
import torch
from torch.utils.data import TensorDataset, DataLoader

# Загружаем токенизатор
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

def tokenize_and_split(texts, max_length=512):
    input_ids_list = []
    attention_mask_list = []
    labels_list = []

    # Токенизация с разбиением на чанки
    for text, label in zip(texts, y_train):  # Мечаем каждый текст с соответствующей меткой
        # Токенизация текста
        encodings = tokenizer(text, truncation=True, padding=False, max_length=max_length)

        # Для каждого чанка создаем отдельную запись
        input_ids_list.append(encodings['input_ids'])
        attention_mask_list.append(encodings['attention_mask'])
        labels_list.extend([label] * len(encodings['input_ids']))  # Повторяем метку для каждого чанка

    return input_ids_list, attention_mask_list, labels_list

# Разделяем данные на чанки
train_input_ids_split, train_attention_mask_split, train_labels_split = tokenize_and_split(X_train.tolist())
test_input_ids_split, test_attention_mask_split, test_labels_split = tokenize_and_split(X_test.tolist())

# Flatten список чанков для всех запросов
train_input_ids_flat = [item for sublist in train_input_ids_split for item in sublist]
train_attention_mask_flat = [item for sublist in train_attention_mask_split for item in sublist]
train_labels_flat = train_labels_split

test_input_ids_flat = [item for sublist in test_input_ids_split for item in sublist]
test_attention_mask_flat = [item for sublist in test_attention_mask_split for item in sublist]
test_labels_flat = test_labels_split

# Проверяем размеры
print(f"train_input_ids_flat: {len(train_input_ids_flat)}")
print(f"train_attention_mask_flat: {len(train_attention_mask_flat)}")
print(f"train_labels_flat: {len(train_labels_flat)}")

print(f"test_input_ids_flat: {len(test_input_ids_flat)}")
print(f"test_attention_mask_flat: {len(test_attention_mask_flat)}")
print(f"test_labels_flat: {len(test_labels_flat)}")

# Убедимся, что размеры совпадают
assert len(train_input_ids_flat) == len(train_attention_mask_flat) == len(train_labels_flat)
assert len(test_input_ids_flat) == len(test_attention_mask_flat) == len(test_labels_flat)

# Преобразуем в PyTorch тензоры
train_input_ids_tensor = torch.tensor(train_input_ids_flat)
train_attention_mask_tensor = torch.tensor(train_attention_mask_flat)
train_labels_tensor = torch.tensor(train_labels_flat)

test_input_ids_tensor = torch.tensor(test_input_ids_flat)
test_attention_mask_tensor = torch.tensor(test_attention_mask_flat)
test_labels_tensor = torch.tensor(test_labels_flat)

# Создаем TensorDataset для тренировки и теста
train_dataset = TensorDataset(train_input_ids_tensor, train_attention_mask_tensor, train_labels_tensor)
test_dataset = TensorDataset(test_input_ids_tensor, test_attention_mask_tensor, test_labels_tensor)

# Создаем DataLoader для батчей
train_dataloader = DataLoader(train_dataset, batch_size=16, shuffle=True)
test_dataloader = DataLoader(test_dataset, batch_size=64)

# Проверка
print(f"Train DataLoader: {len(train_dataloader)} batches")
print(f"Test DataLoader: {len(test_dataloader)} batches")
```

```python
import torch
from torch.utils.data import DataLoader, TensorDataset
from torch.optim import AdamW
from transformers import BertForSequenceClassification, BertTokenizer
from tqdm import tqdm
import shutil
import os
from google.colab import files
import pandas as pd

# --- Подготовка данных (предполагается, что X_train, X_test, y_train, y_test уже существуют) ---

# Загружаем токенизатор
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Преобразуем метки в тензоры, если это pandas.Series
y_train = torch.tensor(y_train.values) if isinstance(y_train, pd.Series) else torch.tensor(y_train)
y_test = torch.tensor(y_test.values) if isinstance(y_test, pd.Series) else torch.tensor(y_test)

# Токенизация текстов
X_train_tokenized = tokenizer(list(X_train), padding=True, truncation=True, return_tensors='pt', max_length=512)
X_test_tokenized = tokenizer(list(X_test), padding=True, truncation=True, return_tensors='pt', max_length=512)

# Проверка соответствия размеров
assert len(X_train_tokenized['input_ids']) == len(y_train)
assert len(X_test_tokenized['input_ids']) == len(y_test)

# Создаем датасеты
train_dataset = TensorDataset(X_train_tokenized['input_ids'], X_train_tokenized['attention_mask'], y_train)
test_dataset = TensorDataset(X_test_tokenized['input_ids'], X_test_tokenized['attention_mask'], y_test)

# DataLoader'ы
train_dataloader = DataLoader(train_dataset, batch_size=16, shuffle=True)
test_dataloader = DataLoader(test_dataset, batch_size=64)

# --- Функции тренировки и валидации ---

def train_epoch(model, dataloader, optimizer, device):
    model.train()
    total_loss = 0

    for batch in tqdm(dataloader, desc="Training"):
        input_ids, attention_mask, labels = [x.to(device) for x in batch]

        optimizer.zero_grad()
        outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    return total_loss / len(dataloader)


def eval_epoch(model, dataloader, device):
    model.eval()
    total_loss = 0

    with torch.no_grad():
        for batch in tqdm(dataloader, desc="Evaluating"):
            input_ids, attention_mask, labels = [x.to(device) for x in batch]

            outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss
            total_loss += loss.item()

    return total_loss / len(dataloader)

# --- Обучение с Early Stopping ---

# Определяем устройство
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Загружаем модель
model = BertForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=2)
model.to(device)

# Оптимизатор
optimizer = AdamW(model.parameters(), lr=1e-5)

# Early stopping параметры
patience = 3
best_val_loss = float("inf")
epochs_without_improvement = 0
epoch = 0

while epochs_without_improvement < patience:
    epoch += 1
    print(f"\nEpoch {epoch}")

    # Тренировка
    train_loss = train_epoch(model, train_dataloader, optimizer, device)

    # Валидация
    val_loss = eval_epoch(model, test_dataloader, device)

    print(f"Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}")

    # Проверка на улучшение
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        epochs_without_improvement = 0
        print("✅ Новая лучшая модель! Сохраняем веса...")
        torch.save(model.state_dict(), "best_model_weights.pth")
    else:
        epochs_without_improvement += 1
        print(f"⚠️ Нет улучшений. Эпох без улучшений: {epochs_without_improvement}/{patience}")

print("✅ Обучение завершено")

# --- Сохранение модели и токенизатора ---

save_path = "my_Llama3_BERT_model"
os.makedirs(save_path, exist_ok=True)

model.save_pretrained(save_path)
tokenizer.save_pretrained(save_path)

print("✅ Модель и токенизатор сохранены в:", save_path)

# Архивируем папку
zip_path = shutil.make_archive(save_path, 'zip', save_path)
print("📦 Архив создан:", zip_path)

# Скачиваем на ПК
files.download(zip_path)
```

```python
import os
import torch
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc, precision_recall_curve, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns

# Указываем папку для сохранения графиков
output_dir = "metrics_outputs"
os.makedirs(output_dir, exist_ok=True)

# Переводим модель в режим инференса
model.eval()

# Генерация предсказаний
y_pred_probs = []  # Для вероятностей
y_pred = []        # Для классов
y_true = []

with torch.no_grad():
    for batch in test_dataloader:
        input_ids, attention_mask, labels = batch

        input_ids = input_ids.to(device)
        attention_mask = attention_mask.to(device)
        labels = labels.to(device)

        # Прогнозы модели
        outputs = model(input_ids, attention_mask=attention_mask)
        logits = outputs.logits

        # Преобразование логитов в вероятности
        probs = torch.softmax(logits, dim=-1)  # Для многоклассовой классификации
        preds = torch.argmax(probs, dim=-1)   # Прогнозируемый класс

        y_pred_probs.extend(probs[:, 1].cpu().numpy())  # Вероятности для класса "1" (при бинарной классификации)
        y_pred.extend(preds.cpu().numpy())             # Предсказанные классы
        y_true.extend(labels.cpu().numpy())            # Истинные метки

# Построение матрицы ошибок
conf_matrix = confusion_matrix(y_true, y_pred)

# Визуализация матрицы ошибок
plt.figure(figsize=(6, 5))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Regular', 'Jailbreak'],
            yticklabels=['Regular', 'Jailbreak'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.savefig(os.path.join(output_dir, "confusion_matrix.png"))  # Сохранение графика
plt.show()

# Генерация отчета по классификации
class_report = classification_report(y_true, y_pred)
print("Classification Report:")
print(class_report)

# ROC-AUC
fpr, tpr, _ = roc_curve(y_true, y_pred_probs)  # Используем вероятности
roc_auc = auc(fpr, tpr)

# Визуализация ROC-кривой
plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, color='b', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.savefig(os.path.join(output_dir, "roc_curve.png"))  # Сохранение графика
plt.show()

# Precision-Recall Curve
precision, recall, _ = precision_recall_curve(y_true, y_pred_probs)  # Используем вероятности

# Визуализация Precision-Recall кривой
plt.figure(figsize=(6, 5))
plt.plot(recall, precision, color='b', lw=2, label='Precision-Recall curve')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc='lower left')
plt.savefig(os.path.join(output_dir, "precision_recall_curve.png"))  # Сохранение графика
plt.show()

# ROC-AUC Score
roc_auc_score_value = roc_auc_score(y_true, y_pred_probs)  # Используем вероятности
print(f"ROC-AUC Score: {roc_auc_score_value:.2f}")
```
